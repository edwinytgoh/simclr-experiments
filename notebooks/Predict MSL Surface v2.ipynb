```python
import os
import sys

import tensorflow as tf

sys.path.insert(0, '/home/goh/Documents/D3M/simclr/tf2/')

from edwin_utils import get_files_and_labels, get_tf_dataset, read_class_label_map
```

```python
msl_folder = '/home/goh/Documents/D3M/Mars_Classification/msl-labeled-data-set-v2.1/'
test_file = os.path.join(msl_folder, 'test-set-v2.1.txt')
img_folder = os.path.join(msl_folder, 'images')
X, num_files, num_classes = get_files_and_labels(img_folder, ext="jpg", metadata_file = test_file)

test_ds = get_tf_dataset(X, "jpg", preprocess=True, width=224, height=224)
```

```python
model_path = '/home/goh/Documents/D3M/mars_simclr_models/pretrained_r152_3x_sk1-2021-06-26_1316-batch32-100epochs-lr0.2'
# model = tf.saved_model.load(model_path)
```

```python
from lars_optimizer import LARSOptimizer

# Initialize optimizer
opt = LARSOptimizer(
    0.2,
    momentum=0.9,
    weight_decay=0.,
    exclude_from_weight_decay=['batch_normalization', 'bias', 'head_supervised'])

km = tf.keras.models.load_model(
    model_path, 
    custom_objects={
        'LARSOptimizer': opt,
        'sparse_softmax_cross_entropy_with_logits_v2': tf.nn.sparse_softmax_cross_entropy_with_logits
    })
```

```python
import numpy as np
```

```python
temp_ds = test_ds.batch(256).prefetch(tf.data.AUTOTUNE)
logits = []
labels = []
for it, (image, label) in enumerate(temp_ds.as_numpy_iterator()):
    logits_t = km(image)
    logits.append(logits_t.numpy())
    labels.append(label)
logits = np.concatenate(logits)
labels = np.concatenate(labels)
predictions = np.argmax(logits, axis=1)
```

```python
logits = np.concatenate(logits)
labels = np.concatenate(labels)
predictions = np.argmax(logits, axis=1)
```

```python
class_mapping = read_class_label_map(os.path.join(msl_folder, 'class_map.csv'))
classes = list(class_mapping.keys())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.metrics
import pandas as pd

y_true = [classes[l] for l in labels]
y_pred = [classes[p] for p in predictions]
cm = sklearn.metrics.confusion_matrix(y_true, y_pred, labels=classes)
cm = pd.DataFrame(cm, index=classes, columns=classes)
cm.index.name = 'Actual'
cm.columns.name = 'Predicted'
fig, ax = plt.subplots(figsize=(14,10))
sns.heatmap(cm, annot=True, cmap='viridis', fmt='.0f')
ax.set_title(f"Finetuned on full training data - overall accuracy: {100*sklearn.metrics.accuracy_score(y_true, y_pred, sample_weight=None):.2f}%")
plt.tight_layout()
# plt.savefig('/home/goh/Documents/D3M/confusion_matrix_mars_full_training.png', transparent=True)
```
